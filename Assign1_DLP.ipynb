```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_wine
```

```python
data=load_wine()
data
```

```python
data1=pd.DataFrame(data['data'],columns=data['feature_names'])
```

```python
data1
```

```python
data1['class']=data['target']
```

```python
data1
```

```python
data1['class']=data1['class'].replace(to_replace=[0,1,2],value=['class_0','class_1','class_2'])
```

```python
data1
```

```python
from sklearn.model_selection import train_test_split
x=data1.iloc[:,0:-1]
y=data1.iloc[:,13:14]
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
xtrain_transform=scaler.fit_transform(xtrain)
xtest_transform=scaler.fit_transform(xtest)
xtrain_transform[:10]
```

```python
xtest_transform
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
model=MLPClassifier(hidden_layer_sizes=(150,100,45),activation='relu',max_iter=300,solver='adam')
```

```python
model.fit(xtrain_transform,ytrain)
```

```python
ypred=model.predict(xtest_transform)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy=accuracy_score(ytest,ypred)
accuracy
```

```python
cm=confusion_matrix(ytest,ypred)
cm
```
